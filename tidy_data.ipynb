## exercise 1:
### Attendance Data

Read the data from the attendance table and calculate an attendance percentage for each student. One half day is worth 50% of a full day, and 10 tardies is equal to one absence.

```python
att = pd.read_csv('untidy-data/attendance.csv')
att = att.rename(columns={'Unnamed: 0':'name'})
att
```

```python
att.head()
att_long = att.melt(id_vars='name', var_name='day', value_name='attendance')
att_long
```

```python
att_long.head()

att_long = att_long.replace({'attendance':{'P':1,'T':.9,'H':.5,'A':0}})
att_long.attendance = att_long.attendance.astype(float)
att_long.groupby('name').attendance.mean()
```

### Cake Recipes

Read the cake_recipes table. This data set contains cake tastiness scores for combinations of different recipes, oven rack positions, and oven temperatures.

Tidy the data as necessary.

Which recipe, on average, is the best? recipe b

Which oven temperature, on average, produces the best results? 275

Which combination of recipe, rack position, and temperature gives the best result? recipe b, bottom rack, 300 degrees

```python
cake = pd.read_csv('untidy-data/cake_recipes.csv')
cake.head()
```

```python
recipe_position = cake['recipe:position'].str.split(':', expand=True)
recipe_position.columns=['recipe','position']
recipe_position.head()
```

```python
cake2 = pd.concat([cake, recipe_position], axis=1).drop(columns='recipe:position')
cake2.head()
```

```python
cake2 = cake2.pivot(index=['recipe'], columns=['position'])
cake2
```

```python
cake2
```

```python
cake3 = cake2.T
cake3=cake3.reset_index()
cake3 = cake3.rename(columns={'level_0':'temp'})
cake3
```
